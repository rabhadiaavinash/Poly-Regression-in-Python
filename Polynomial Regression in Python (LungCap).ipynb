```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
data = pd.read_excel("c:\Users\Kishore\Desktop\Datasets\LungCap.xls")
```

```python
data.head()
```

```python
X = data.loc[:,['Height']].values
Y = data.loc[:,'LungCap'].values
```

```python
X = X.reshape(-1, 1)
```

```python
X
```

```python
plt.scatter(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
Lreg1 = LinearRegression()
Lreg1.fit(X,Y)
```

```python
plt.scatter(X,Y, color = "red")
plt.plot(X , Lreg1.predict(X) , color = "blue")
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
poly_reg = PolynomialFeatures(degree =3)
```

```python
X_poly = poly_reg.fit_transform(X)
```

```python
X_poly
```

```python
df = pd.DataFrame(X_poly)
```

```python
df.head()
```

```python
Lreg2 = LinearRegression()
Lreg2.fit(X_poly,Y)
```

```python
plt.scatter(X,Y , color ="red")
plt.plot(X , Lreg2.predict(X_poly), color = "blue")
```
